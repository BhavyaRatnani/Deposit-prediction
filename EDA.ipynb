```python
data.select_dtypes(include=['object']).columns
categ_var=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact','poutcome', 'y']
```

```python
data.select_dtypes(include=['int64']).columns
nume_var=['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
```

```python
data.isnull().sum()*100/len(data)
```

```python
sns.countplot(x=data['y'])
```

```python
for i in categ_var:
  plt.figure(figsize=(20,8))
  ax=sns.countplot(x=data[i])
  #label_values(ax, spacing=-15)
  plt.show()
```

```python
for i in categ_var:
    plt.figure(figsize=(20,8))
    ax=sns.catplot(x='y', col=i, kind='count', data= data)
    #label_values(ax, spacing=-15)
plt.show()
```

```python
for i in categ_var:
    print(data.groupby(['y',i]).size())
```

```python
for i in nume_var:
  plt.figure(figsize=(10,4))
  ax=sns.distplot(x=data[i])
  plt.xlabel(i)
  plt.show()
```

```python
for i in nume_var:
  plt.figure(figsize=(20,10))
  #ax=sns.boxplot(x="y", y= data[i], data=data)
  ax=sns.boxplot(data[i])
  plt.xlabel(i)
  plt.show()
```
